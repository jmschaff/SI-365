```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import math
import sys
import time
import csv
```

```python
N = 1000000
Nbin = 50

mu_x = 95
sigma_x = .001

mu_y = -95
sigma_y = 1000

alpha = 0.05

x, y = np.random.multivariate_normal([mu_x, mu_y], [[sigma_x**2,alpha*sigma_x*sigma_y], [alpha*sigma_x*sigma_y,sigma_y**2]],N).T

plt.hist2d(x, y, bins=Nbin, range=None, normed=False, weights=None, cmin=None, cmax=None, hold=None)
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.title('2D Histogram')
plt.show()


#Uncomment the lines below to save in a csv file
both = np.array([x,y],float)
save = both.T

with open('14tuesday.csv', 'w') as f:
     writer = csv.writer(f)
     writer.writerows(save)
```

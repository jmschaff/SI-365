###Import necessary libraries

```python
import random
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline  
```

###Random walk without using numpy

```python
## pure Python way

position = 0
walk = [position]
steps = 100

for i in xrange(steps):
    step = 1 if random.randint(0,1) else -1
    position += step
    walk.append(position)

plt.plot(walk)
plt.show()
```

###Random walk using numpy

```python
# Uncomment the commented the lines to see what the variables return
nsteps = 1000
draws = np.random.randint(0, 2, size=nsteps)
# print draws[0:10]
steps = np.where(draws > 0, 1, -1)
# print steps[0:10]
walk = steps.cumsum()
# print walk[0:10]

print "walk.min = {0}".format(walk.min())
print "walk.max = {0}".format(walk.max())
# print "first crossing time above {0} = {1}".format(10, (np.abs(walk) >= 10).argmax())

plt.plot(walk)
plt.show()
```
